```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PowerTransformer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('concrete_data.csv')
df.head()
```

```python
x = df.drop(columns=['Strength'])
y = df['Strength']
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)
r2_score(y_test,y_pred)
```

```python
pt = PowerTransformer()

x_train_pt = pt.fit_transform(x_train + 0.00000001)
x_test_pt = pt.transform(x_test + 0.00000001)
```

```python
lr = LinearRegression()

lr.fit(x_train_pt,y_train)

y_pred2 = lr.predict(x_test_pt)
r2_score(y_test,y_pred2)
```

```python
pt = PowerTransformer(method='box-cox')

x_train_pt2 = pt.fit_transform(x_train + 0.00000001)
x_test_pt2 = pt.transform(x_test + 0.00000001)
```

```python
lr = LinearRegression()

lr.fit(x_train_pt2,y_train)

y_pred2 = lr.predict(x_test_pt2)
r2_score(y_test,y_pred2)
```
